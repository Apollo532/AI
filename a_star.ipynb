WHAT IS A STAR ALGORITHM?

A* Search Algorithm is a simple and efficient search algorithm that can be used to find the optimal path between two nodes in a graph. It will be used for the shortest path finding. It is an extension of Dijkstra's shortest path algorithm (Dijkstra's Algorithm).

WHY IS IT USED?

A* Search Algorithm is a simple and efficient search algorithm that can be used to find the optimal path between two nodes in a graph. It will be used for the shortest path finding. It is an extension of Dijkstra's shortest path algorithm (Dijkstra's Algorithm).
It is more efficient than BFS and DFS as it uses both of them

The time complexity of A* depends on the heuristic. In the worst case of an unbounded search space, the number of nodes expanded is exponential in the depth of the solution (the shortest path) d: O(bd), where b is the branching factor (the average number of successors per state).

HOW TO USE?

```python

from queue import PriorityQueue

# Define the graph as a dictionary of nodes and their neighbors
graph = {
    'A': {'B': 5, 'C': 10},
    'B': {'D': 6},
    'C': {'D': 8, 'E': 10},
    'D': {'E': 5},
    'E': {}
}

# Define the heuristic function to estimate the distance from a node to the goal node
heuristic = {
    'A': 14,
    'B': 12,
    'C': 8,
    'D': 4,
    'E': 0
}

# A* search function
def astar_search(start, goal):
    frontier = PriorityQueue()
    frontier.put((0, start))
    came_from = {}
    cost_so_far = {}
    came_from[start] = None
    cost_so_far[start] = 0
    # IF THE GIVEN GRAPH IS NOT EMPTY
    while not frontier.empty():
      # THE LOWEST PRIORITY ON THE QUEUE WILL BE CHOOSEN ( WHICH IS OF THE LOWEST COST)
        current = frontier.get()[1]
        #IF THE CHOOSEN NODE IS THE GOAL NODE, THEN WE STOP

        if current == goal:
            break

        for next_node in graph[current]:
            new_cost = cost_so_far[current] + graph[current][next_node]
            if next_node not in cost_so_far or new_cost < cost_so_far[next_node]:
                cost_so_far[next_node] = new_cost
                priority = new_cost + heuristic[next_node]# f(n)= g(n)+h(n)
                frontier.put((priority, next_node))
                came_from[next_node] = current

# Reconstruct the path from the start to the goal node
    path = [goal]
    while path[-1] != start:
        path.append(came_from[path[-1]])
    path.reverse()

# Calculate the shortest distance between each pair of nodes in the path
    distances = {}
    for i in range(len(path) - 1):
        current_node = path[i]
        next_node = path[i + 1]
        distance = graph[current_node][next_node]
        distances[(current_node, next_node)] = distance

    return path, distances

```

```python
# Example usage
start_node = 'A'
goal_node = 'E'
path, distances = astar_search(start_node, goal_node)
print(' -> '.join(path))
print(distances)
```
