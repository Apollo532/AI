WHAT IS SINGLE LAYER PERCEPTRON?

A single layer perceptron (SLP) is a feed-forward network based on a threshold transfer function. SLP is the simplest type of artificial neural networks and can only classify linearly separable cases with a binary target (1 , 0).


WHY TO USE SINGLE LAYER PERCEPTRON?

It helps to organize the given input data. A perceptron is a neural
network unit that does a precise computation to detect features in the input data. Perceptron is mainly used to classify the data into two parts. Therefore, it is also known as Linear Binary Classifier.

HOW TO USE IT?

```python
import numpy as np
from matplotlib import pyplot as plt
```

```python

X = np.array([[1.6,0.6, 0], [1.2, 0, 1], [0, 1.2, 1], [0.3, 0.6, 1]])
y = np.array([[1], [1], [0], [0]])
```

```python

def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

```python
np.random.seed(0)
w = np.random.randn(3, 1)
```

```python
epochs = 10
lr = 0.1
for epoch in range(epochs):
    # Forward pass
    z = np.dot(X, w)
    a = sigmoid(z)
    print("Before error computation: ", a)

    # Compute the error
    error = y - a
    print("ERRORS: ",error)

    # Backward pass
    w += lr * np.dot(X.T, error * a * (1 - a))
    print("After Backward pass:" ,w)
```

```python
test_input = np.array([[1.5, 0.2, 0]])
result = sigmoid(np.dot(test_input, w))
print(result)
```

```python
plt.plot(X,sigmoid(X))
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='coolwarm', edgecolor='black', label='Input Data')
```
